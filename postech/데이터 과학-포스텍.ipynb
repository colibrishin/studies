데이터 과학
- 통계학 + 데이터 마이닝 + 빅 데이터 + 딥러닝 + 인공지능

- 데이터 분석
    - 데이터 큐레이션
    - 데이터 추출, 변환
    - 데이터 시각화
    - 통계모형, 인공지능, t-검정, 회귀분석, 머신러닝

- 데이터 분석 -> 새로운 결과 -> 인사이트

- 통계적 개념과 지식 (샘플링, 확률분포, 가설검정, p-value)
- 데이터를 다룰 수 있는 기술 (큐레이션)
- 데이터의 요약된 정보 전달 기술 (시각화 - 공간지도분석, 다차원 그래픽)
- 데이터 윤리, 데이터 보안
- 데이터 도메인에 대한 지식과 분석능력 (현실 문제의 해결능력)
